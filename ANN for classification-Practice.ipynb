```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
import warnings
warnings.simplefilter("ignore")
```

```python
import tensorflow 
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
iris_data=load_iris()
```

```python
iris_data
```

```python
print(iris_data.data[:5])
```

```python
print(iris_data.target[:5])
```

```python
X=iris_data.data
```

```python
X
```

```python
X.shape
```

```python
y=iris_data.target.reshape(-1,1)
y
```

```python
encoder=OneHotEncoder(sparse=False)
y=encoder.fit_transform(y)
print(y)
```

```python
x_train,x_test,y_train, y_test=train_test_split(X,y,test_size=0.20,random_state=11)
```

```python
x_train.shape
```

# Model Building

# 1.building Architec of the model

####2. Model Prediction

####3. model Evaluate

```python
model=Sequential()
#we have to arrange Sequential that so we ar e using the equential
```

```python
model.add(Dense(10,input_shape=(4,),activation='relu'))
#Dense is used to craete  layer of a neuron
#10-Neurons no ..
#input_shape=No of  input featuers(no of input columns)
#activation==all 10 Neurons are activate in 'Relu'(transformation oparation )
```

```python
#craeting second  hiddeen Layer
model.add(Dense(10,activation='relu'))
#no need to mention the Input shape beacuse alredy we are mention on our 1st Layer ..now its taking data from the First Layer
#We can Create N no of Layers
```

```python
#craeting output Layer
model.add(Dense(3,activation='softmax'))
#3===no of calsess we have on our Target variable
#Softmax==Activation function for the output layer 
```

```python
optimizer=Adam(lr=0.001)
#Adam one kind of optimizer
#lr=Learning Rate(Grediant Decent)
```

```python
#compiling the Model
model.compile(optimizer,loss='categorical_crossentropy',metrics=['accuracy'])
#Optimizer== variable taken for optimize on above cell
#loss=loss fuction there is several loss function is there ,but Here We are dealing with the Catagorical data that so we taken â†’ categorical_crossentropy
#Metics==evaluate Matrics ,
#Why Accuracy?===hence we know this a classification problem that so we are using the Accuracy
```

```python
#Saw the Model of our Model
model.summary()
```

# Parm# No of weights ==  (No of input shape +1)* No of Neurorns

# Model building End

```python
#epochs== no of ittretion
model.fit(x_train,y_train,batch_size=5,epochs=40)
```

```python
result=model.evaluate(x_test,y_test)
```
